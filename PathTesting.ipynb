```python
@dataclass 
class sph: 
    N = 300
    Nth = 5
    dth = 4 / 5
    Nr = 8
    dr = 3 / 8
    inpx = np.expand_dims(np.linspace(0, 2*np.pi, N), 1)

self = sph()
```

```python
# np.sin(self.inpx).shape
# r.shape
# self.inpx.shape
# theta.shape
# print((a * np.sin(i*self.inpx)).shape)
np.linspace(1, 2, self.N, axis=-1).shape
np.linspace(np.pi/2, 2*np.pi, self.N).shape
```

```python
size = 100

# Initialize vectors 
theta_start = np.random.rand(1)
theta = np.linspace(theta_start, theta_start + 2*np.pi, self.N)
r = np.ones(theta.shape)
# r = np.linspace(0, 1, num=self.N)
r = np.squeeze(r) 

print(theta.shape)

# Randomize theta 
for i in range(self.Nth): 
    a = (np.random.rand(1)-.5) * self.dth
    # theta += a * np.sin(i * self.inpx)
    theta = theta + (a * np.sin(i * self.inpx))

print((theta + (np.sin(i * self.inpx))).shape)

# finalize theta 
theta = theta * np.sign(np.random.rand(1)-.5)  # Randomly flip the sign 
theta = theta[:-1]  # Cut off the last place

print(f"R shape pre {r.shape}")

# Randomize radius 
for i in range(self.Nr): 
    a = (np.random.rand(1)-.5) * self.dr
    r = r + (a * np.squeeze(np.sin(i*self.inpx)))
    
print(f"R shape post {r.shape}")

# Finalize radius
r = r[:-1]

# print(r.size)

# Swap into coordinate system 
x = size * r * np.squeeze(np.cos(theta))
y = size * r * np.squeeze(np.sin(theta))

x = np.squeeze(x)
y = np.squeeze(y)
```

```python
save_flag = True
def save_it(s): 
    if save_flag: 
        plt.savefig(s + ".png")
    else: 
        plt.show()

fig = plt.figure()
gs = fig.add_gridspec(2, 3)

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[:, 1:])

# plt.show()

ax3.plot(x, y, '.')
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.set_title("Target Path") 
ax3.grid()
# save_it("TargetPath")
        
ax1.plot(theta)
ax1.set_ylabel("Theta")
ax1.set_xlabel("Time Step")
# plt.title("Angle of path per step")
# plt.grid()
# save_it("TargetTheta")

ax2.plot(r)
ax2.set_xlabel("Time Step")
ax2.set_ylabel("Radius")
# ax2.grid()
# save_it("TargetRadius")

# Fix it 
fig.tight_layout()
save_it("TargetPath")
```

```python
import pygame
from pygame.locals import *
```

```python
clr = (255, 0, 0)

print("initializing...") 
pygame.init()
 
fps = 60
fpsClock = pygame.time.Clock()
 
width, height = 640, 480
screen = pygame.display.set_mode((width, height))

# Game loop.
print("Looping...") 
i = 0
while True:
    screen.fill((0, 0, 0))
    
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            # sys.exit()
  
    # Update.
    i += 1 
    if len(x) <= i: 
        i = 0 
    
    # Draw.
    pygame.draw.circle(screen, clr, (x[i] + width/2, height/2-y[i]), 10)
      
    pygame.display.flip()
    fpsClock.tick(fps)
```
